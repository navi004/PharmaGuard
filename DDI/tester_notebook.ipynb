```python
df3 = pd.read_csv("parsed_data/drugs.csv")
print(df3.shape)
df3.head()
```

```python
# count of nulls per column and print each column on its own line with percentage
null_counts = df3.isnull().sum()
for col, cnt in null_counts.items():
    pct = (cnt / len(df3) * 100) if len(df3) else 0
    print(f"{col}: {cnt} nulls ({pct:.2f}%)")

# total number of nulls in the dataframe
print("Total nulls:", null_counts.sum())
```

```python
# get raw unique values (each cell as-is)
raw_unique = df3['categories'].unique()
print(f"Raw unique count: {len(raw_unique)}")
print(raw_unique)

# get exploded unique values (split semicolon-separated category lists, trim, drop empties)
exploded = (
    df3['categories']
    .dropna()
    .astype(str)
    .str.split(';')   # adjust delimiter if needed
    .explode()
    .str.strip()
)
exploded = exploded[exploded != '']
unique_categories = exploded.unique()
print(f"\nExploded unique count: {len(unique_categories)}")
print(sorted(unique_categories))

# optional: frequency of each category
print("\nTop categories by count:")
print(exploded.value_counts().head(20))
```
